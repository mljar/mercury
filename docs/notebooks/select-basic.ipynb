# Basic usage

Display `Select` in the sidebar, and use its `value` in the main view.

```python
fruit = Select(label="Choose fruit", 
             choices=["Apple ðŸŽ", "Banana ðŸŒ", "Cherry ðŸ’"])
```

```python
print(f"You selected: {fruit.value}")
```

## Control position

Change `position` to display slider in main view or bottom.

```python
algo = Select(label="Select algorithm", 
              choices=["Random Forest", "Neural Networks", "Cant tell you"],
              position="inline")
```

```python
print(f"My favourite algo is {algo.value}")
```

### Display in the bottom

You can add widgets in the bottom too!

```python
llm = Select(label="LLM",
             choices=["LLama", "Qwen", "GPT OSS", "Mistral"],
             position="bottom")
```

```python
print(f"You are using {llm.value}")
```
