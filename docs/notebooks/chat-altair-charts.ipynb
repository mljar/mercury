```python
import mercury as mr
import pandas as pd
import altair as alt
```

```python
sales_df = pd.DataFrame({
    'date': ['2023-01-01', '2023-01-08', '2023-01-15', '2023-01-22'],
    'region': ['North', 'South', 'East', 'West'],
    'product': ['A', 'B', 'A', 'C'],
    'units_sold': [10, 15, 7, 12],
    'unit_price': [20, 25, 19, 22]
})
sales_df['total_sales'] = sales_df['units_sold'] * sales_df['unit_price']
```

```python
chat = mr.Chat()
```

```python
prompt = mr.ChatInput()
```

```python
if prompt.value:
    user_msg = mr.Message(prompt.value, role="user")
    chat.add(user_msg)

    response_msg = mr.Message("**Altair** plot based on sales data", role="assistant", emoji='ðŸ¤–')
    with response_msg:
        # Create data for plotting
        df = sales_df.copy()
        df['date'] = pd.to_datetime(df['date'])

        # Create Altair line plot
        chart = alt.Chart(df).mark_line(point=True).encode(
            x=alt.X('date:T', title='Date'),
            y=alt.Y('total_sales:Q', title='Total Sales'),
            tooltip=['date:T', 'total_sales:Q']
        ).properties(
            width=400,
            height=250,
            title='Total Sales Over Time'
        )

        display(chart)
    
    chat.add(response_msg)
```
