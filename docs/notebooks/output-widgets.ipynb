```python
import pandas as pd
import mercury as mr

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "score": [82, 91, 77]
})

csv_data = df.to_csv(index=False)

mr.Download(
    data=csv_data,
    filename="scores.csv",
    mime="text/csv",
    label="Download CSV"
)
```

```python
mr.Indicator(value="123", label="Users", delta=5.4)
```

```python
mr.Indicator([
    mr.Indicator(value="123", label="Users", delta=5.4),
    mr.Indicator(value="98%", label="Accuracy", delta=-1.2),
    mr.Indicator(value="1.3s", label="Latency")
     ])
```
