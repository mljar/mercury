```python
import pandas as pd
import mercury as mr

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "score": [82, 91, 77]
})

csv_data = df.to_csv(index=False)

mr.Download(
    data=csv_data,
    filename="scores.csv",
    mime="text/csv",
    label="Download CSV"
)
```
