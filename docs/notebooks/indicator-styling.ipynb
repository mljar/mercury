```python
import mercury as mr
from IPython.display import display
```

```python
mr.Markdown("## 1. Default Built-in Indicator\nBy default, the indicator provides a clean, neutral card.")

# Checklist Item 1
mr.Indicator(value="1,024", label="Total Active Users")
```

```python
mr.Markdown("---\n## 2. Custom Styles Indicators\nYou can customize the look of your indicators using semantic variants or explicit hex colors.")

# Checklist Item 2: Variants
ind1 = mr.Indicator(value="98%", label="ACCURACY", variant="success")
ind2 = mr.Indicator(value="340ms", label="LATENCY", variant="warning")
ind3 = mr.Indicator(value="$12k", label="REVENUE", variant="danger")

display(ind1, ind2, ind3)
```

```python
# Checklist Item 2: Explicit Colors
mr.Indicator(
    value="$12,340",
    label="CUSTOM PURPLE",
    accent_color="#7c3aed",
    border_color="#ddd6fe",
    value_color="#111827",
    label_color="#6b7280"
)
```
