```python
import mercury as mr
```

# Output widgets

They help you control how your results are displayed to the user. Let's build something beautiful.

### Columns

```python
cols = mr.Columns(3, border="1px solid lightgray")
```

```python
with cols[0]:
    print("First column")
with cols[1]:
    print("Second column")
with cols[2]:
    print("Third column")
```

### Expander

```python
details = mr.Expander(label="Experiment details")
```

```python
with details:
    print("Many interesting details are here indeed.")
```

### Tabs

```python
tabs = mr.Tabs(labels=["Input", "Output", "Settings"])
```

```python
with tabs[0]:
    print("Input data goes here")

with tabs[1]:
    print("Output results go here")

with tabs[2]:
    print("List all settings")
```
