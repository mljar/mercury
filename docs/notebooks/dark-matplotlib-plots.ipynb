# Dark Theme Matplotlib Plots in Mercury

This tutorial shows how to create beautiful dark theme matplotlib plots and use them in Mercury apps.

Dark theme plots are perfect for:
- Dashboards with dark backgrounds
- Presentations and reports
- Reducing eye strain in low-light environments

## Setup

Install the required libraries:

```python
# Install dependencies
# pip install mercury matplotlib numpy
```

```python
import mercury as mr
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

## Mercury App Configuration

```python
app = mr.App(
    title="Dark Matplotlib Plots",
    description="Tutorial: Dark theme matplotlib plots in Mercury"
)
```

## Available Dark Styles

Matplotlib comes with several built-in dark styles:

```python
# View all available dark styles
dark_styles = [s for s in plt.style.available if 'dark' in s.lower()]
print('Available dark styles:', dark_styles)
```

## Method 1: Using dark_background style

```python
# Select chart type
chart_type = mr.Select(
    value="Line",
    choices=["Line", "Bar", "Scatter"],
    label="Chart Type"
)

# Select dark style
dark_style = mr.Select(
    value="dark_background",
    choices=["dark_background", "Solarize_Light2"],
    label="Dark Style"
)
```

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Apply selected dark style
with plt.style.context(dark_style.value):
    fig, ax = plt.subplots(figsize=(10, 5))

    if chart_type.value == "Line":
        ax.plot(x, y1, color='#7C3AED', linewidth=2, label='sin(x)')
        ax.plot(x, y2, color='#38BDF8', linewidth=2, label='cos(x)')
        ax.plot(x, y3, color='#34D399', linewidth=2, label='sin(x)cos(x)')
    elif chart_type.value == "Bar":
        categories = ['A', 'B', 'C', 'D', 'E']
        values = np.random.randint(10, 100, 5)
        colors = ['#7C3AED', '#38BDF8', '#34D399', '#F59E0B', '#EC4899']
        ax.bar(categories, values, color=colors)
    elif chart_type.value == "Scatter":
        scatter_x = np.random.randn(100)
        scatter_y = np.random.randn(100)
        ax.scatter(scatter_x, scatter_y, color='#7C3AED', alpha=0.6, s=50)

    ax.set_title(f'Dark Theme {chart_type.value} Chart', fontsize=14, pad=15)
    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    if chart_type.value == "Line":
        ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

## Method 2: Custom Dark Theme

You can also create a fully custom dark theme:

```python
# Custom dark theme settings
custom_dark = {
    'figure.facecolor': '#0f0f0f',
    'axes.facecolor': '#1a1a1a',
    'axes.edgecolor': '#333333',
    'axes.labelcolor': '#ffffff',
    'text.color': '#ffffff',
    'xtick.color': '#6B7280',
    'ytick.color': '#6B7280',
    'grid.color': '#333333',
    'grid.alpha': 0.5
}

with matplotlib.rc_context(custom_dark):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Plot 1 - Line chart
    axes[0].plot(x, y1, color='#A78BFA', linewidth=2)
    axes[0].fill_between(x, y1, alpha=0.2, color='#A78BFA')
    axes[0].set_title('Custom Dark Line Chart', color='#ffffff')
    axes[0].grid(True)

    # Plot 2 - Bar chart
    categories = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
    values = [23, 45, 32, 67, 54]
    bars = axes[1].bar(categories, values,
                       color=['#A78BFA', '#38BDF8', '#34D399',
                              '#F59E0B', '#EC4899'])
    axes[1].set_title('Custom Dark Bar Chart', color='#ffffff')
    axes[1].grid(True, axis='y')

    plt.tight_layout()
    plt.show()
```

## Summary

You can create dark matplotlib plots in Mercury using:

1. **Built-in styles** - `plt.style.use('dark_background')` for quick dark themes
2. **Custom themes** - Use `matplotlib.rc_context()` for full control over colors

Both methods work perfectly with Mercury widgets for interactive dark dashboards.
