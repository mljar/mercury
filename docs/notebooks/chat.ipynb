# Chat example ðŸ’¬

This is simple echo chat, we simply repeat the prompt. This is a good starting point to develop your chat or AI agent.

---

```python
chat = mr.Chat()
```

```python
prompt = mr.ChatInput()
```

```python
if prompt.value:
    user_msg = mr.Message(prompt.value, role="user")
    chat.add(user_msg)

    response_msg = mr.Message(f"Echo: {prompt.value}", role="assistant")
    chat.add(response_msg)
```
