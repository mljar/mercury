```python
import mercury as mr
```

```python
chat = mr.Chat()
```

```python
prompt = mr.ChatInput()
```

```python
if prompt.value:
    # user message
    user_msg = mr.Message(prompt.value, role="user")
    chat.add(user_msg)

    # assistant message (echo)
    response_msg = mr.Message(
        f"Echo: {prompt.value}",
        role="assistant",
        emoji="ðŸ¤–",
    )
    chat.add(response_msg)
```
