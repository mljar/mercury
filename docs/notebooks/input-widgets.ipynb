# Mercury Input Widgets

In this web app we show case Mercury Input Widgets. They can be used to quickly turn your notebook into interactive web app.

```python
btn = mr.Button(label="Click me!")
check = mr.CheckBox(label="I agree")
check2 = mr.CheckBox(label="Consent", appearance="box")
color = mr.Select(
    label="Choose color",
    choices=["Red", "Green", "Blue"]
)
fruits = mr.MultiSelect(
    label="Choose fruits",
    choices=["Apple ðŸŽ", "Banana ðŸŒ", "Cherry ðŸ’"],
    value=["Banana ðŸŒ", "Cherry ðŸ’"]
)
rows = mr.NumberInput(label="Rows", value=5)
threshold = mr.Slider(
    label="Threshold",
    min=0,
    max=10,
    value=3
)
name = mr.TextInput(
    label="Enter your name",
    value="Aleksandra"
)
uploader = mr.UploadFile(
    label="Upload CSV",
    multiple=False
)
```

### Button

```python
if btn.value:
    print("You clicked button!")
else:
    print("Button not clicked ...")
```

### Checkbox

```python
print(f"Check boxes: {check.value} {check2.value}")
```

### MultiSelect

```python
print(f"MultiSelect fruits: {fruits.value}")
```

### NumberInput

```python
print(f"Rows: {rows.value}")
```

### Select

```python
print(f"Selected color: {color.value}")
```

### Slider

```python
print(f"Threshold value: {threshold.value}")
```

### TextInput

```python
print(f"Hi {name.value} ðŸ‘‹")
```
