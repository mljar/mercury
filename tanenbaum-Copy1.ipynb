```python
import mercury as mr
from IPython.display import HTML
from dotenv import load_dotenv
import html as _html
import os
_ = load_dotenv(".ollama.env") 
```

```python
from ollama import Client

client = Client(
    host="https://ollama.com",
    headers={'Authorization': 'Bearer ' + os.environ.get('OLLAMA_API_KEY')}
)
```

```python
messages = [{
    'role': 'system',
    'content': 'Respond with SVG ONLY'
}]
```

```python
left, right = mr.Columns(2)
```

```python
with right:
    right.clear()
    preview, code = mr.Tabs(["Preview", "Code"])
```

```python
with left:
    chat = mr.Chat()
```

```python
prompt = mr.ChatInput()
```

```python
if prompt.value:
    chat.add(mr.Message(prompt.value))
    messages += [{'role': 'user', 'content': prompt.value}]
    response = client.chat(model="gpt-oss:20b", messages=messages)
    
    messages += [
      {
        'role': response.message.role,
        'content': response.message.content,
      },
    ]

    with preview:
        preview.clear()
        display(SVG(response.message.content))
    with code:
        code.clear()
        print(response.message.content)
    
```
